```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
data.groupby('event_source').size()
```

```python
data.event_source.value_counts()[:2].sort_values(inplace=False).plot('bar',title ='Fuente Eventos',color = ['#E0B627','#27E0C4']);
```

```python
data.groupby('host').size().sort_values(inplace=False)
```

```python
datos = data.host.unique().tolist()
colors = ['darkgreen', 'lightgreen', 'gold']
name = data.groupby(['host']).size().reset_index(name='count')
percent = name['count'].tolist()
plt.pie(percent, labels = datos, colors = colors)
plt.legend(loc="lower center")
plt.title('Host Accedido')
```

```python
data.groupby('accept_language').size().sort_values(inplace=False)
```

```python
data.accept_language.value_counts()[:10].sort_values(inplace=False).plot('bar',title ='Idiomas con mayor interacción',color = ['#E0B627','#27E0C4','#2D27E0','#35E027']);
```
